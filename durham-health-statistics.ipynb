### SECTION 1: Data Loading

```python
# Loading Data as a CSV using Pandas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

health_df = pd.read_csv('data/health.csv')
key_indicators_df = pd.read_csv('data/key_indicators.csv')
```

### SECTION 2: Exploratory Data Analysis (EDA)
---


```python
# Step 1: Display the shape and first 10 rows of the health dataset to understand its structure
print("Health Dataset shape:")
print(health_df.shape)

print("\nDataset preview:")
health_df.head(10)
```

```python
# Display the shape and first 10 rows of the key indicators dataset to understand its structure
print("Health Dataset shape:")
print(key_indicators_df.shape)

print("\nDataset preview:")
key_indicators_df.head(10)
```

```python
# Step 2: Check the data types of each column
print("\nData types:")
health_df.dtypes
```

```python
print("\nData types:")
key_indicators_df.dtypes
```

```python
# Step 3: Check for missing values
print("\nMissing values:")
print(health_df.isnull().sum())
```

```python
print("\nMissing values:")
print(key_indicators_df.isnull().sum())
```

```python
# Step 4: Generate basic summary statistics for numerical columns
print("\nSummary statistics for numerical columns:")
print(health_df.describe())
```

```python
print("\nSummary statistics for numerical columns:")
print(key_indicators_df.describe())
```

##### Note(s):
1. All the categorical columns have the values we would expect to be stored in them.

```python
# Step 6: Check for duplicate rows
print("\nNumber of duplicate rows:")
print(health_df.duplicated().sum())
print(key_indicators_df.duplicated().sum())
```

##### Note(s):
1. There are no duplicate rows in the dataset.

```python
# Visualize missing data as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(key_indicators_df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Data Heatmap - Key Indicators")
plt.show()
```

```python
# Male Life Expectancy
plt.subplot(1, 2, 1)
sns.histplot(key_indicators_df["Life Expectancy Male"], kde=True, bins=20, color="skyblue")
plt.title("Distribution of Life Expectancy (Male)")
plt.xlabel("Life Expectancy (Years)")
plt.ylabel("Frequency")

# Female Life Expectancy
plt.subplot(1, 2, 2)
sns.histplot(key_indicators_df["Life Expectancy Female"], kde=True, bins=20, color="salmon")
plt.title("Distribution of Life Expectancy (Female)")
plt.xlabel("Life Expectancy (Years)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# --- Correlation Heatmap ---
plt.figure(figsize=(12, 8))
corr_matrix = key_indicators_df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", vmin=-1, vmax=1)
plt.title("Correlation Heatmap - Key Indicators")
plt.show()
```

```python
# --- Boxplot of Obesity Levels ---
plt.figure(figsize=(12, 6))
sns.boxplot(x="Obesity", data=key_indicators_df, color="lightgreen")
plt.title("Boxplot of Obesity Levels")
plt.xlabel("Obesity (%)")
plt.show()
```

```python
# --- Scatter Plot: Smoking vs Life Expectancy (Male) ---
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="Smoking",
    y="Life Expectancy Male",
    data=key_indicators_df,
    hue="MUNICIPALITY_NAME",
    palette="tab10",
    edgecolor="w"
)
plt.title("Smoking vs Life Expectancy (Male)")
plt.xlabel("Smoking (%)")
plt.ylabel("Life Expectancy (Years)")
plt.legend([],[], frameon=False)  # Hides legend for a cleaner display
plt.show()
```
